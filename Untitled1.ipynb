```python
def trial(deck_dic):
    deck_map = []
    for d, s in deck_dic.keys():
        max_num = deck_dic.get((d, s))
        for i in range(max_num+1):
            i = str(i) # need convert numeric into strings to join
            join_result = '/'.join((d, i, s))
            deck_map.append(join_result)
    return deck_map

deck_map = trial(deck_dic)
```
